```python
import numpy as np
import pandas as pd

import scipy.stats as stats

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PowerTransformer
from sklearn.metrics import r2_score
from sklearn.model_selection import cross_val_score


from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv("/content/concrete_data.csv")
```

```python
df.head(3)
```

```python

df.isnull().sum()
```

```python
df.describe()
```

```python
x=df.drop(columns=['Strength'])
y=df.iloc[:,-1]
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_tr,y_tr)
y_pred=lr.predict(x_te)
r2_score(y_te,y_pred)
```

```python
np.mean(cross_val_score(lr,x,y,scoring='r2'))
```

```python
for col in x_tr.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_tr[col])
  plt.subplot(122)
  stats.probplot(x_tr[col],dist='norm',plot=plt)
  plt.title(col)
  plt.show()
```

```python
pt = PowerTransformer(method='box-cox')
x_tr_pt=pt.fit_transform(x_tr+0.00001)
x_te_pt=pt.transform(x_te+0.00001)
pd.DataFrame({'cols':x_tr.columns,'box-cox_lambdas':pt.lambdas_})
```

```python
lr=LinearRegression()
lr.fit(x_tr_pt,y_tr)
y_pred2=lr.predict(x_te_pt)
r2_score(y_te,y_pred2)
```

```python
x_trans=pt.fit_transform(x+0.00001)
np.mean(cross_val_score(lr,x_trans,y,scoring='r2'))
```

```python
x_tr_pt=pd.DataFrame(x_tr_pt,columns=x_tr.columns)
x_te_pt=pd.DataFrame(x_te_pt,columns=x_te.columns)
```

```python
for col in x_tr_pt.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_tr[col])
  plt.title(col)
  plt.subplot(122)
  sns.distplot(x_tr_pt[col])
  plt.title(col)
  plt.show()
```

```python
pt1=PowerTransformer() # default yeo johnson
x_tr_pt2=pt1.fit_transform(x_tr)
x_te_pt2=pt1.transform(x_te)
pd.DataFrame({'cols':x_tr.columns,'yeo_johnson_lambdas':pt1.lambdas_})
```

```python
lr=LinearRegression()
lr.fit(x_tr_pt2,y_tr)
y_pred2=lr.predict(x_te_pt2)
r2_score(y_te,y_pred2)
```
