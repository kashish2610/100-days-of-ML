```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic.csv')
```

```python
df
```

# number

```python
fig=df['number'].value_counts().plot(kind='bar')
```

```python
df['number_num']=pd.to_numeric(df['number'],errors='coerce',downcast='integer')
```

```python
df['number_cat']=np.where(df['number_num'].isnull(),df['number'],np.nan)
```

```python
df
```

# cabin

```python
 df['Cabin_num']=df['Cabin'].str.extract('(\d+)') # capture numeric values
df['cabin_cat']=df['Cabin'].str[0]
```

```python
df
```

```python
# extract last bit of ticket as no.
df['tic_num']=df['Ticket'].apply(lambda s: s.split()[-1])
df['tic_num']=pd.to_numeric(df['tic_num'],errors='coerce',downcast='integer')
# extract 1st part of tic as cat
df['tic_cat']=df['Ticket'].apply(lambda s: s.split()[0])
df['tic_cat']=np.where(df['tic_cat'].str.isdigit(),df['tic_cat'],np.nan)
```
