```python
import numpy as np
import  pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv("/content/placement (1).csv")
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
x_tr,x_te,y_tr,y_te= train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
x_trf=lr.fit(x_tr,y_tr)
```

```python

plt.scatter(df['cgpa'],df['package'])
plt.plot(x_te,lr.predict(x_te),color='red')
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
y_pr=lr.predict(x_te)
```

```python
y_te.values
```

```python
print("MAE", mean_absolute_error(y_te,y_pr))
```

```python
print("MSE", mean_squared_error(y_te,y_pr))
```

```python
print("RMSE",np.sqrt(mean_squared_error(y_te,y_pr)))
```

```python
print("r2",r2_score(y_te,y_pr))
r2=r2_score(y_te,y_pr)
```
