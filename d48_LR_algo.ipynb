```python
class MeraLR:
  def __init__(self):
    self.m=None
    self.b=None

  def fit(self,x_tr,y_tr): # m= $((x_i - x_mean)(y_i-y_mean))/$(x_i-x_mean)^2
    print(x_tr.shape)
    num=0
    den=0
    for i in range(x_tr.shape[0]):
      num = num + (x_tr[i] - x_tr.mean())*(y_tr[i] - y_tr.mean())
      den = den + (x_tr[i] - x_tr.mean())*(x_tr[i] - x_tr.mean())
    self.m= num/den
    self.b = y_tr.mean() - self.m*(x_tr.mean())
    print(self.m)
    print(self.b)

  def predict(self,x_te):
    print(x_te)
    return self.m*x_te + self.b
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr,x_te,y_tr,y_te= train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr=MeraLR()
lr.fit(x_tr,y_tr)
```
