```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
```

```python
df=pd.read_csv('/content/train (1).csv')
```

```python
df
```

```python
df=df.drop(columns=['PassengerId','Name','Ticket','Cabin'])
```

```python
df.head(2)
```

```python
x_tr,x_te,y_tr,y_te=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2,random_state=42)
```

```python
y_tr.head(2)
```

```python
df.isnull().sum()
```

```python
# Apply imutation
si_age=SimpleImputer()
si_embarked=SimpleImputer(strategy='most_frequent')
```

```python
x_tr_age=si_age.fit_transform(x_tr[['Age']])
x_tr_embarked=si_embarked.fit_transform(x_tr[['Embarked']])
x_te_age=si_age.transform(x_te[['Age']])
x_te_embarked=si_embarked.transform(x_te[['Embarked']])
```

```python
# onehotenc
ohe_sex=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
ohe_emb=OneHotEncoder(sparse_output=False,handle_unknown='ignore')
x_tr_sex=ohe_sex.fit_transform(x_tr[['Sex']])
x_tr_emb=ohe_emb.fit_transform(x_tr_embarked)
x_te_sex=ohe_sex.transform(x_te[['Sex']])
x_te_emb=ohe_emb.transform(x_te_embarked)
```

```python
x_tr_emb
```

```python
x_tr_age.shape
```

```python
x_tr_sex.shape
```

```python
x_tr_rem=x_tr.drop(columns=['Sex','Age','Embarked'])
x_te_rem=x_te.drop(columns=['Sex','Age','Embarked'])
```

```python
x_te_rem.shape
```

```python
x_tr_trans=np.concatenate((x_tr_rem, x_tr_age, x_tr_sex, x_tr_emb),axis=1)
x_te_trans=np.concatenate((x_te_rem,x_te_age,x_te_sex,x_te_emb),axis=1)
```

```python
x_tr_trans
```

```python
 clf = DecisionTreeClassifier()
 clf.fit(x_tr_trans,y_tr)
```

```python
y_pred =clf.predict(x_tr_trans)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_tr, y_pred)
```
