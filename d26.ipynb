```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/customer.csv')
```

```python
df.sample(4)
```

```python
df=df.iloc[:,2:]
df
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te= train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
y_tr
```

```python
# categories gives priorites to data like poor <avg<good and sch<ug<pg
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_tr)

# Transform data
x_tr=oe.transform(x_tr)
x_te=oe.transform(x_te)
```

```python
x_tr
```
