# Fetching different pages

```python
import pandas as pd
import requests

data = pd.DataFrame()  # Initialize an empty DataFrame

for i in range(1, 5):
    url =  f"https://api.themoviedb.org/3/movie/top_rated?api_key=6b5579c43ed73840efebe09f924ad0ed&language=en-US&page={i}"
    headers = {"accept": "application/json"}
    response = requests.get(url, headers=headers)

    temp_df = pd.DataFrame(response.json()['results'])[['id','title','overview','popularity','vote_average','vote_count']]  # Convert API response to DataFrame

    data = pd.concat([data, temp_df], ignore_index=True)  # Use pd.concat() instead of append()

#print(data.head())  # Print first few rows to check the output'
data.head()
```
