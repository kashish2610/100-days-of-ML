```python
#[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]
```

```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
# @title we need top 3 dictionary of every movie and only need name key
movies['cast'][0]
```

```python
def convert1(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if counter !=3:
      l.append(i['name'])
      counter+=1
    else:
      break
  return l
```

```python
movies['cast']= movies['cast'].apply(convert1)
```

```python
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
# @title we need only job director
def convert2(obj):
  l=[]
  counter=0
  for i in ast.literal_eval(obj):
    if i['job']=='Director':

      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(convert2)
```

```python
movies.head(1)
```

```python
# @title convert string to list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
# @title 'Sam Worthing' ---> 'SamWorthing
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head(4)
```

```python
df=movies[['movie_id','title','tags']]
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df['tags']=df['tags'].apply(lambda x:x.lower())
```

# Vectorization


```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vec=cv.fit_transform(df['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
ps.stem('loving')
```

```python
df['tags'][0]
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver james cameron')
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
def recomd(movie):
  movie_index=df[df['title']==movie].index[0]
  distance=similarity[movie_index]
  movie_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_list:
    print(df.iloc[i[0]].title)
  return
```
